## Creando un paquete
Son una forma de crear entornos virtuales sin afectar paquetes instalados globalmente u otros programas

## Crear un entorno virtual
Instalar virtual env con:
```
    sudo apt install python3.8-ven
```
Despues ejecutar:
```
    python3 -m venv env
```
Activar el entorno virtual ejecutando:
```
    source env/bin/activate
```
